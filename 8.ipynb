```python
X = df.iloc[:,[0,1]]
X
```

```python
Y = df.iloc[:,[2]]
Y
```

```python
from sklearn.model_selection import train_test_split
Xtrain,Xtest,Ytrain,Ytest = train_test_split(X,Y,random_state = 0,test_size = 0.25)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
Xtrain = scaler.fit_transform(Xtrain)
```

```python
Xtest = scaler.fit_transform(Xtest)
```

```python
from sklearn.linear_model import LogisticRegression
regressor = LogisticRegression()
```

```python
regressor.fit(Xtrain,Ytrain)
```

```python
y_test_predict = regressor.predict(Xtest)
```

```python
y_test_predict
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cm = confusion_matrix(Ytest,y_test_predict)
cm
```

```python
acc = accuracy_score(Ytest,y_test_predict)*100
acc
```

```python
tp = cm[0][0]
tn = cm[1][1]
fn = cm[0][1]
fp = cm[1][0]
total = tp+tn+fn+fp
total
```

```python
error_rate =((fp+fn)/total)
print('Total error rate is {}'.format(error_rate))
```

```python
from sklearn.metrics import classification_report
classification_report(Ytest,y_test_predict)
```

```python
print('                      Classification Report:')
print(' ')
print(classification_report(Ytest,y_test_predict))
```
