```python
for k,v in df.items():
    q1 = v.quantile(0.25)
    q3 = v.quantile(0.75)
    iqr = q3 - q1
    aa = (v <= (q1 - 1.5*iqr))
    bb = (v >= (q3 + 1.5*iqr))
    v_col = v[aa | bb]
    perc = np.shape(v_col)[0]*100.0/np.shape(df)[0]
    print("Column %s outliers = %.2f%%" % (k,perc))
```

```python
data = df[~(df['MEDV']>=50.0)]
print(data.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
col = data.columns
```

```python
df_scaled = scaler.fit_transform(data)
df_scaled = pd.DataFrame(df_scaled,columns = col)
print("Scaled Dataset Using MinMaxScaler")
df_scaled
```

```python
df_scaled = df_scaled.dropna()
df_scaled.shape
```

```python
X = pd.DataFrame(np.c_[df_scaled['LSTAT'],df_scaled['RM']],columns = ['LSTAT' , 'RM'])
```

```python
Y = df_scaled['MEDV']
```

```python
from sklearn.model_selection import train_test_split
Xtrain,Xtest,Ytrain,Ytest = train_test_split(X,Y,test_size = 0.4, random_state=42)
print(Xtrain)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
from sklearn.linear_model import LinearRegression
lin_model = LinearRegression()
```

```python
lin_model.fit(Xtrain,Ytrain)
```

```python
Y_train_predict = lin_model.predict(Xtrain)
rmse = np.sqrt(mean_squared_error(Ytrain,Y_train_predict))
r2  = r2_score(Ytrain,Y_train_predict)

print("The model performance for training set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")
```

```python
Y_test_predict = lin_model.predict(Xtest)
rmse = np.sqrt(mean_squared_error(Ytest,Y_test_predict))
r2  = r2_score(Ytest,Y_test_predict)

print("The model performance for testing set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")
```
